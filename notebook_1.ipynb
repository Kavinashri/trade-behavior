```python
import pandas as pd

root = "/content/ds_Kavinashri"

# load the two datasets
trader = pd.read_csv(f"{root}/csv_files/trader_data.csv", low_memory=False)
sentiment = pd.read_csv(f"{root}/csv_files/fear_greed.csv")

# preview first few rows
print("Trader data:")
display(trader.head())

print("\nSentiment data:")
display(sentiment.head())
```

```python

trader['Timestamp IST'] = pd.to_datetime(trader['Timestamp IST'], errors='coerce')
sentiment['date'] = pd.to_datetime(sentiment['date'], errors='coerce')

trader['date'] = trader['Timestamp IST'].dt.date
sentiment['date'] = sentiment['date'].dt.date

merged = trader.merge(sentiment[['date','classification']], on='date', how='left')

print("Merged dataset shape:", merged.shape)
merged.head()
```

```python
# Profitability grouped by sentiment
profit_stats = merged.groupby('classification')['Closed PnL'].agg(['count','mean','median','std']).reset_index()
print("Profitability by Sentiment")
print(profit_stats)

# Save results
profit_stats.to_csv("/content/ds_Kavinashri/csv_files/profit_stats_by_sentiment.csv", index=False)
```

```python
# If leverage column exists in dataset
if 'Leverage' in merged.columns:
    lev_stats = merged.groupby('classification')['Leverage'].agg(['count','mean','median']).reset_index()
    print("\nLeverage by Sentiment")
    print(lev_stats)

    # Save
    lev_stats.to_csv("/content/ds_Kavinashri/csv_files/lev_stats_by_sentiment.csv", index=False)
else:
    print("No leverage column found in dataset.")
```

```python
vol = merged.groupby(['classification','Side'])['Size USD'].sum().reset_index().sort_values(['classification','Size USD'], ascending=[True,False])
print("\nTrade Volume by Sentiment and Side")
print(vol)

# Save
vol.to_csv("/content/ds_Kavinashri/csv_files/volume_by_sentiment_side.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(8,5))
sns.boxplot(x='classification', y='Closed PnL', data=merged)
plt.title('Closed PnL by Market Sentiment')
plt.savefig("/content/ds_Kavinashri/outputs/pnl_by_sentiment.png")
plt.close()

vol_pivot = merged.groupby(['classification','Side'])['Size USD'].sum().unstack(fill_value=0)
vol_pivot.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title('Trade Volume by Sentiment and Side')
plt.ylabel('Total USD Volume')
plt.savefig("/content/ds_Kavinashri/outputs/volume_by_sentiment_side.png")
plt.close()

print("Plots saved in outputs/ folder")
```

```python
from fpdf import FPDF

pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)

# Title
pdf.set_font("Arial", 'B', 14)
pdf.cell(200, 10, "Trader Behavior vs Market Sentiment Analysis", ln=True, align="C")
pdf.set_font("Arial", size=12)
pdf.cell(200, 10, "Candidate: Kavinashri", ln=True, align="C")
pdf.ln(10)

# Objective
pdf.multi_cell(0, 10, "1. Objective\n\nThe goal of this assignment is to explore how trader behavior "
    "(profitability, trading volume, buy/sell activity) correlates with the Fear & Greed Index.\n"
    "By aligning trader actions with sentiment, we can identify whether crowd psychology influences trading outcomes.\n")

# Data Sources
pdf.multi_cell(0, 10, "2. Data Sources\n- Trader Data: Execution price, trade size, side (BUY/SELL), Closed PnL.\n"
    "- Sentiment Data: Daily Fear & Greed index values with classifications.\n"
    "- Data Merge: Both datasets were joined using the date column.\n")

# Analysis - Profitability
pdf.multi_cell(0, 10, "3. Analysis\n\n3.1 Profitability by Sentiment\n- Boxplot of Closed PnL vs Sentiment.\n"
    "- Result: Greed days show higher average profitability, while Fear days show wider spread with some negative PnL.\n")

pdf.multi_cell(0, 10, "Figure 1: Closed PnL by Market Sentiment")
pdf.image("/content/ds_Kavinashri/outputs/pnl_by_sentiment.png", w=150)
pdf.ln(10)

# Analysis - Trade Volume
pdf.multi_cell(0, 10, "3.2 Trade Volume by Sentiment & Side\n- Aggregated USD trade size by Sentiment and Side.\n"
    "- Result: Greed -> Higher BUY activity, Fear -> More SELL activity, lower volume overall.\n")

pdf.multi_cell(0, 10, "Figure 2: Trade Volume by Sentiment and Side")
pdf.image("/content/ds_Kavinashri/outputs/volume_by_sentiment_side.png", w=150)
pdf.ln(10)

# Insights
pdf.multi_cell(0, 10, "4. Key Insights\n1. Greed periods show higher volumes and more BUYs.\n"
    "2. Fear periods have lower activity and more SELLs.\n"
    "3. Profitability is generally higher in Greed phases.\n"
    "4. Sentiment index can act as a leading indicator.\n")

# Deliverables
pdf.multi_cell(0, 10, "5. Deliverables\n- Notebook: notebook_1.ipynb\n- CSV outputs in csv_files/\n"
    "- Charts in outputs/\n- Final Report: ds_report.pdf\n")

# Conclusion
pdf.multi_cell(0, 10, "6. Conclusion\nThe analysis confirms that trader behavior is influenced by sentiment.\n"
    "Traders are more active and profitable during Greed, while Fear reduces volume and increases selling.\n"
    "This highlights the usefulness of combining quantitative signals with market psychology.\n")

# Save PDF
pdf.output("/content/ds_Kavinashri/ds_report.pdf")
print(" Final report with charts saved as ds_report.pdf")
```
